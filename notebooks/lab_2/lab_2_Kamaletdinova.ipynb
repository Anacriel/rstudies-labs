##  We are going to test 'NoPodsWeight' and 'PodsWeight' more.

#### Custom t-test function

```r
t_test <- function(sample_1, sample_2) {
    result <- list(t_stat = 0, p_value = 0)
    var_1 <- var(sample_1)
    var_2 <- var(sample_2)
    
    len_1 <- length(sample_1) 
    len_2 <- length(sample_2)
    
    freedom_degr <- length(sample_1) + length(sample_2) - 2
    var_p <- sqrt(var_1 * var_1 + var_2 * var_2)
    
    result$t_stat <- (mean(sample_1) - mean(sample_2)) / (var_p * sqrt(1 / len_1))
    result$p_value <- 2 * pt(abs(result$t_stat), df = freedom_degr, lower.tail = FALSE)
    
    return(result)
}
```
