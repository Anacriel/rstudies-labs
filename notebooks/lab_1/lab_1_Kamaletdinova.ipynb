## Merge sorts from both files

#### Read tables in data frames

```r
df = read.table('test_data1.txt', header=TRUE)
head(df)
```

```r
sapply(df[,-1], median)
```

```r
df_chrom = read.table('test_data2.csv', sep=';', header=TRUE)
head(df_chrom)
```

```r
df_chrom_transp <- as.data.frame(t(df_chrom[,-1]))
colnames(df_chrom_transp) <- df_chrom$ID

sorts_col <- rownames(df_chrom_transp)
rownames(df_chrom_transp) <- NULL

df_chrom_transp <- cbind(sorts_col, df_chrom_transp)
names(df_chrom_transp)[1]<-paste("Sample")
```

```r
head(df_chrom_transp)
```

```r
df_binded = merge(df, df_chrom_transp, by="Sample")
head(df_binded)
```

## Replace NA values if some of SNPs are not defined for a sort

```r
sort_snps_len = nrow(df_chrom)
```

#### Modify original table adding the table with replaced NAs in SNPs

```r
features_len = ncol(df)
sort_snps_len = nrow(df_chrom)

df_snps <- df_binded[, (features_len + 1):ncol(df_binded)]
df_snps[(rowSums(is.na(df_snps)) < sort_snps_len) & is.na(df_snps)] <- 0

df_binded <- data.frame(df_binded[, 1:features_len], df_snps)
df_binded
```
